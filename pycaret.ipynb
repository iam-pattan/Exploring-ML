### Binary-Class Classification

```python
#### Classification Example for Binary
diabetes_df=get_data("diabetes")
```

```python
diabetes_df['Class variable'].value_counts()
```

```python
#### data preprocessing and analysis
from pycaret.classification import *
clf=setup(data=diabetes_df,target='Class variable')
```

```python
#### Comparing models
compare_models()
```

### Mulit-Class Classification

```python
#### Classification Example for Multi-class
poker_df=get_data("poker")
```

```python
poker_df['CLASS'].value_counts()
```

```python
#### data preprocessing and analysis
from pycaret.classification import *
clf=setup(data=poker_df,target='CLASS', use_gpu = True)
```

```python
#### Comparing models
compare_models()
```

### Regression
